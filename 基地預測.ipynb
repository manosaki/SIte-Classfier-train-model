```python
import pandas as pd
pd.set_option('display.max_rows', None)

#只顯示row15，column10
#pd.set_option('display.max_rows', 15)
#pd.set_option('display.max_columns', 10)

# scikit-learn 會有些deprecation warning, 為了顯示漂亮, 我刻意地忽略掉
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("/drive/My Drive/Colab Notebooks/基地環境1.csv",
          encoding = "utf-8")
df
```

# 資料預處理

```python
df['標籤'] = df['標籤'].astype('category')
saved_map = { cat:df['標籤'].cat.categories.get_loc(cat) for cat in df['標籤'].cat.categories}
saved_map
```

```python
# 定義好等等我們要對所有內容做的split
def split_content(content):
  return " ".join(jieba.cut(content))
def process_contents(df):
  df['內容'] = df['內容'].apply(split_content)
  df['內容'] = df['內容'].str.replace('\r', '')
  df['內容'] = df['內容'].str.replace('\n', '')
  df['標籤'] = df['標籤'].astype('category')
  return df
```

```python
import jieba
df = process_contents(df)
df
pre_label = df['標籤']
pre_label
```

```python
df['標籤'] = df['標籤'].cat.codes
df
```

```python
test_df = pd.read_csv("/drive/My Drive/Colab Notebooks/基地環境_test1.csv", 
                      encoding = "utf-8")
test_df = process_contents(test_df)
test_df
label = test_df['標籤']
```

```python
test_df['標籤'] = test_df['標籤'].replace(saved_map)
test_df
```

# 建立詞向量

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
vec = TfidfVectorizer()
bag = vec.fit_transform(df['內容'])
print("維度:", len(vec.get_feature_names()))
clf = MultinomialNB(alpha = 0.001)
clf.fit(bag, pre_label)
```

# 建立詞袋並訓練

```python
test_bag = vec.transform(test_df['內容'])
test_bag
```

# 預測

```python

from sklearn.metrics import accuracy_score
test_bag = vec.transform(test_df['內容'])
predict = clf.predict(test_bag)
print("預測:", list(predict))
print("正確標籤:", list(label))
print("Naive-Bayes 正確率: ", accuracy_score(label, predict) * 100, "%")
```
